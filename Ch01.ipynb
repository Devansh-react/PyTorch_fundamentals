# Mathematical operations

```python
x = torch.rand(2,2)
x
```

```python
# addition
print(x + 2)
# substraction
print(x - 2)
# multiplication
print(x * 3)
# division
print(x / 3)
# int division
print((x * 100)//3)
# mod
print(((x * 100)//3)%2)
# power
print(x**2)
```

Elementwise operation

```python
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
```

```python
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** b)
print(a % b)
```

```python
c = torch.tensor([-1.111,0.29929292,3.9999,-4.1290])
c
```

```python
print(c.abs())
print(c.neg())
print(c.round())
print(c.ceil())
print(c.floor())
print(c.clamp(min=0,max=3))
```

```python
e = torch.randint(size=(2,3),low=7,high=10,dtype=torch.float32)
e
```

```python
print(torch.sum(e))
# sum along col
print(torch.sum(e,dim=0))
print(torch.sum(e,dim=1))
print(torch.mean(e,dim=0))
print(torch.mean(e,dim=1))
print(torch.median(e,dim=0))
print(torch.min(e))
print(torch.std(e))
print(torch.var(e))

#  tell the position of max/min element
print(torch.argmax(e))
print(torch.argmin(e))
```

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)

print(f)
print(g)
```

```python
# matrix multiplcation
torch.matmul(f, g)
```

```python
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])

# dot product
torch.dot(vector1, vector2)
```

```python
torch.transpose(f,0,1)
# swaping the 0 with 1
# torch.det(g)
```

```python
i = torch.randint(size=(2,3), low=0, high=10)
j = torch.randint(size=(2,3), low=0, high=10)

print(i)
print(j)
```

```python
# greater than
i > j
# less than
i < j
# equal to
i == j
# not equal to
i != j
# greater than equal to
```

```python
k = torch.randint(size=(2,3), low=5, high=10, dtype=torch.float32)
k
```

```python
print(torch.sigmoid(k))
print(torch.softmax(k,dim=0))
print(torch.relu(k))
```
